ML

```python
# Step 1: Data Preparation
# Aggregate features per user: Total Transaction Value and Average Redeemed Points
user_aggregates = transactions.groupby('CustomerID').agg({
    'TransactionValue': ['sum', 'mean'],                # Ensure 'TransactionValue' matches the column in your dataset
    'TransactionRedeemedPoints': ['sum', 'mean']        # Ensure 'TransactionRedeemedPoints' matches the column in your dataset
}).reset_index()

# Rename columns for easier access
user_aggregates.columns = ['UserID', 'TotalTransactionValue', 'AverageTransactionValue', 'TotalRedeemedPoints', 'AverageRedeemedPoints']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram
# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(user_aggregates[['TotalTransactionValue', 'AverageTransactionValue', 'TotalRedeemedPoints', 'AverageRedeemedPoints']])
```

```python
# Step 3: Dimensionality Reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
reduced_data = pca.fit_transform(scaled_data)
```

```python
#K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_data)
user_aggregates['KMeans_Cluster'] = kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
# Evaluate K-Means
silhouette_kmeans = silhouette_score(scaled_data, user_aggregates['KMeans_Cluster'])
davies_bouldin_kmeans = davies_bouldin_score(scaled_data, user_aggregates['KMeans_Cluster'])

print(f'Silhouette Score for K-Means: {silhouette_kmeans}')
print(f'Davies-Bouldin Index for K-Means: {davies_bouldin_kmeans}')
```

```python
# Gaussian Mixture Model
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=5)
user_aggregates['GMM_Cluster'] = gmm.fit_predict(reduced_data)

# Evaluate GMM
silhouette_gmm = silhouette_score(reduced_data, user_aggregates['GMM_Cluster'])
davies_bouldin_gmm = davies_bouldin_score(reduced_data, user_aggregates['GMM_Cluster'])
```

```python
print(f'Silhouette Score for GMM: {silhouette_gmm}')
print(f'Davies-Bouldin Index for GMM: {davies_bouldin_gmm}')
```
